```python
import gffpandas.gffpandas as gffpd
import pandas as pd
```

```python
annotation = gffpd.read_gff3('YPS1009.gff')
```

```python
annotation.header
```

```python
annotation.df
```

```python
combined_df = annotation.filter_feature_of_type(['gene'])
```

```python
combined_df.df
```

```python
attr_to_columns = combined_df.attributes_to_columns()
```

```python
attr_to_columns.columns
```

```python
DF = attr_to_columns[['seq_id', 'start', 'end', 'Name',  'Alias',  'Note', 'Ontology_term',
       'copy_num_ID', 'coverage', 'curie', 'display',
       'extra_copy_number', 'gene', 'low_identity', 'orf_classification',
       'partial_mapping', 'sequence_ID', 'valid_ORFs']]
```

```python
# This code just convert the chromosome numbering to integer and remove mitochondrial genes.
Latin_to_arabic = {'I': 1, 'II':2, 'III':3, 'IV':4, 'V':5, 'VI':6, 'VII':7, 'VIII':8, 'IX':9, 'X':10,
                  'XI':11, 'XII':12, 'XIII':13, 'XIV':14, 'XV':15, 'XVI':16, 'Mito': 'mitoch', 'MANY': 'MANY'}

seq_id = DF['seq_id']
ch = []
contig = []
for i in seq_id:
    chrom, cont = i.split('_')
    contig.append(cont)
    chrom = chrom.replace('chr', '')
    if '-' in chrom:
        chrom, e = chrom.split('-')
    ch.append(Latin_to_arabic[chrom])
DF['Chromosome'] = ch
DF['Contig'] = contig
```

```python
note = DF['Note']
dis = DF['display']
clean = []
display = []
for n in note:
    clean.append(n.replace('%20', ' '))
for d in dis:
    display.append(str(d).replace('%20', ' '))
DF['Note'] = clean
DF['display'] = display
```

```python
DF.columns
```

```python
DF = DF[['Name', 'Chromosome', 'Contig', 'start', 'end', 'Name', 'Alias', 'Note', 'Ontology_term',
       'coverage', 'curie', 'display', 'extra_copy_number',
       'gene', 'low_identity', 'orf_classification', 'partial_mapping',
       'sequence_ID', 'valid_ORFs']]
```
