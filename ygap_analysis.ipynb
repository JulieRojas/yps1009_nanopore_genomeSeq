```python
df_trna = pd.read_csv('YPS1009.trna-v2.csv', index_col=0)
```

```python
df_trna = df_trna[['Strand', 'LowestCoordinate', 'HighestCoordinate', 'S288C_chrom',
       'YGAP_contigNumber', 'GeneShortName', 'GeneCoordinates', 'GeneType']]
```

```python
# This code just convert the chromosome numbering to integer and remove mitochondrial genes.
Latin_to_arabic = {'I': 1, 'II':2, 'III':3, 'IV':4, 'V':5, 'VI':6, 'VII':7, 'VIII':8, 'IX':9, 'X':10,
                  'XI':11, 'XII':12, 'XIII':13, 'XIV':14, 'XV':15, 'XVI':16, 'Mito': 'mitoch', 'MANY': 'MANY'}

ch_s = df_trna['S288C_chrom']
ch = []

for c in ch_s:
    chrom = c.replace('chr', '')
    if '-' in chrom:
        chrom, e = chrom.split('-')
    ch.append(Latin_to_arabic[chrom])
df_trna['Chromosome'] = ch
```

```python
trna_s288 = pd.read_csv('tRNA_S288C.tsv', sep = '\t')
```

```python
trna_s288.head()
```

```python
import regex as re

tname = list(trna_s288['tRNA Gene > Sgd Alias'])
new = []
for t in tname:
    m = re.search(r"[t][A-Z]\([A-Z]{3}\)",t)
    if m: 
        a = m.group(0)
        a = a.replace('(', '-')
        a = a.replace(')', '')
    else:
        a = None
    new.append(a)
```

```python
trna_s288['GeneShortName'] = new
```
